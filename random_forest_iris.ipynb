RANDOM FOREST


```python
# Importação das bibliotecas
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Carrega o dataset iris
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
iris
```

## x = independentes
## y = dependente

```python
#divide os dados em conjunto de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) #70% train e 30% teste
```

```python
#criar um modelo de regressão logistica
model = RandomForestClassifier(n_estimators=50, min_samples_leaf=3, max_features=2, oob_score=True)

#n_estimators=50 => número de árvores

#trina o modelo
model.fit(X_train, y_train)
```

```python
# Faz previsões no conjunto de teste
y_pred = model.predict(X_test)

# Calcula a acurácia do modelo
acuracy = accuracy_score(y_test, y_pred)
print("Acurácia:", acuracy)
```

```python
#Média do resultados dos treinamentos das árvores
print(model.oob_score_)
```

```python
#matriz de decisáo de cada árvore
print(model.oob_decision_function_)
```

```python
#grau de importancia de cada atributo
print(model.feature_importances_)
```
